## pre-Processing 

### 1.converting to lowercase

```python
df["review"]=df["review"].str.lower()
```

### removing urls

```python
import re 

sample_text = "abc is the word, abc"  # abc =>xyz

new_text = re.sub("abc","xyz",sample_text)

new_text
```

```python
def remove_url(text):   # \S =>non-white space charcter , \s=> whites space
    text=re.sub(r"http\S+" ,"",text)   # (pattern,repl,string) eg-[http://www.google.com]
    return text

    
df["review"]=df["review"].apply(remove_url)
    
```

### 3.Removing punctuations

```python
def remove_panctuations(text):
    text =re.sub(r"[^A-Za-z0-9\s]","",text) # panctuations=>(A-z,a-z,0-9 /s)
    return text 

df["review"]=df["review"].apply(remove_panctuations)
```

```python
df.head()
```

### 4. Removing HTML

```python
def remove_html(text):
    text =re.sub(r"<.*?>","",text) # 
    return text 

df["review"]=df["review"].apply(remove_html)
```

### 5. Removing the Stopwords

```python
import nltk 

nltk.download("punkt")
nltk.download("stopwords")
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
sample_text ="I like coding in python "
tokens =word_tokenize (sample_text)
```

```python
tokens
```

```python
df.head()
```

```python
def remove_stopwords(text):
    tokens=word_tokenize(text)
    stop_words = stopwords.words("english")

    for word in tokens:
        if word in stop_words:
            text =text.replace(word, "")
    return text

df["review"]=df["review"].apply(remove_stopwords)
```

### Stemming

```python
# running -> run
# played->play
# porterStrimming 

from nltk.stem import PorterStemmer 
```

```python
def stemming(text):
    ps =PorterStemmer()
    stemmed_words =[]

    tokens =word_tokenize(text)
    for token in tokens:
        stemmed_token = ps.stem(token)
        stemmed_words.append(stemmed_token)

    return " ".join(stemmed_words)

df["review"]=df["review"].apply(stemming)
```

```python
from torch.utils.data import TensorDataset,DataLoader
import torch 
import numpy as np 
```

```python
X_train = X_train.toarray()
X_test = X_test.toarray()
```

```python
X_train
```

```python

train_set =TensorDataset(
    torch.from_numpy(X_train).float(),
    torch.from_numpy(y_train.values).float()
)
test_set =TensorDataset(
    torch.from_numpy(X_test).float(),
    torch.from_numpy(y_test.values).float()
)
```

```python
train_loader = DataLoader(train_set,shuffle =True,batch_size=64)
test_loader = DataLoader(test_set,shuffle =True,batch_size=64)
```

### Build RNN

```python
import torch.nn as nn 
import torch.optim  as optim
```

```python
class RNN(nn.Module):
    def __init__(self,input_size,hidden_size=128,num_layers=1):
        super().__init__()

        self.hidden_size = hidden_size
        self.num_layers = num_layers

        # RNN layers
        self.rnn =nn.RNN(input_size,hidden_size,num_layers,batch_first=True) # batch_first perameter => change its shape

        # fully connected layers
        self.fc = nn.Linear(hidden_size ,1)

    def forward(self,x):
        # (optional  => shape(num of layers ,batch_size,hidden_layers))
        h0 = torch.zeros(self.num_layers , x.size(0),self.hidden_size)

        out, _=self.rnn(x,h0)
         # 1st value = hidden state of all the timesteps =>(batch ,seq_len,hidden size)
        # 2 nd value = final hidden state of last timestep

        out =self.fc(out[:,-1,:])
        return out 
```

```python
input_size =X_train.shape[1]

model =RNN(input_size)

criterion = nn.BCELoss() # binary cross entropy loss 
optimizer = optim.Adam(model.parameters())
```

### Training RNN

```python
# unsqueeze => add 1 extra dimenstion in data 
# squeeze =>remove 1 dimension 
```

```python
epochs =10
for epoch in range(epochs):
    model.train()

    for Xb,yb in train_loader:
        optimizer.zero_grad()

        Xb =Xb.unsqueeze(1)  # ADD singleton direction 

        outputs =model(Xb)  # (batch_size =1)

        outputs=torch.sigmoid(outputs.squeeze())  # (batch_size, ) activation function used it for probability outputs

        loss =criterion(outputs,yb)  # compute loss 
        loss.backward() # backpropagation
        optimizer.step() # weight update

    print(f"epoch ={epoch+1}/{epochs} and loss ={loss.item()}")
        
        
```

```python
# Evalution

model.eval()
with torch.no_grad():
    correct_vals =0
    tot_vals =0

    for Xb,yb,in test_loader:
        Xb =Xb.unsqueeze(1)

        outputs =model(Xb)
        predicted = (torch.sigmoid(outputs.squeeze()) >0.5).float()

        tot_vals += yb.size(0)
        correct_vals += (predicted == yb).sum().item()

    print(f"accuracy ={correct_vals / tot_vals *100}")
```
